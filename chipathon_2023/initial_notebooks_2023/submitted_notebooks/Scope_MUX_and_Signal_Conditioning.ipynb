**_Abstract_** - The programmable gain amplifier for the scope signal conditioning can be implemented as a non-inverting amplifier using a folded cascode op amp. The gain can be varied using a digitally controlled resistor bank for the feedback resistor. The scope inputs can be buffered using a source follower stage and multiplexing can be achieved by switching the bias current of that stage. 

**_Key words_** - Programmable Gain, Non-Inverting Amplifier, Folded Cascode, Resistor Bank, Source Follower. 


## I. Introduction

Precise gains ranging from 0.25 to 8 in steps with a factor of 2 suggest the use of a (fully differential) op amp in negative feedback as a non-inverting amplifier. The feedback resistor can be digitally controlled using a resistor bank to achieve variable gain. The fully differential op amp can be implemented using the folded cascode topology, chosen for its high (open loop) gain, to improve the DC accuracy of the programmable gain, and its larger swing limits. The inputs to the non-inverting amplifier can be buffered (to avoid loading the scope inputs) through a source follower stage. Multiple channels can be realized by connecting the outputs of multiple source follower stages to the input of a single non-inverting amplifier, and multiplexing can be achieved by turning on the bias current of only a single source follower stage at a time.
