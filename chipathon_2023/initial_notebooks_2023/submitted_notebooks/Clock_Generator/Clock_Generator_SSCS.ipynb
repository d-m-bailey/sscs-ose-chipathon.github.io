**_Abstract_** - Clock generators are essential components in many integrated circuits (ICs) as they provide timing signals for proper operation and coordination of various subsystems. In designing Lab bench on chip, we propose a clock generator circuit using PLL and digital logic gates to generate a stable and accurate clock signal for  any submodule on Lab bench on a chip.  

**_Key words_** - PLL; Clock Generator; Jitter; Phase Noise; Programmable Frequency Divider


## I. Introduction

In modern electronic systems, Analog-to-Digital Converters (ADC) and Digital-to-Analog Converters (DAC) are commonly used to interface between the analog and digital domains. These interface electronics require a stable and precise clock signal to ensure reliable and accurate conversion of analog signals to digital and vice versa. Therefore, a clock generator circuit is an essential component of any ADC or DAC system or any sub-module of a system-on-chip(SoC) for signal processing.

The purpose of this proposal is to outline the design and implementation of a clock generator module for clocking myriad sub-modules in a Lab Bench on Chip which require flexibility in clock frequency with Low jitter while consuming very little power. 



## II. Implementation Details of your Idea



Clock_generator1.jpg

The block diagram depicted above highlights the building blocks of the proposed Charge Pump Based Clock Generator Circuit. The details of the sub-blocks are listed below.

Input Divider: It is responsible for dividing the frequency of the input signal to the PLL, which is typically a high-frequency reference signal, down to a lower frequency that can be easily processed by the other components of the PLL.

Phase Frequency Detector: It compares the phase and frequency of the reference signal with the feedback signal, and generates an error signal that is used to adjust the frequency and phase of the PLL's output signal. The PFD plays a critical role in the PLL's ability to lock onto and track the frequency and phase of the reference signal, ensuring stable and accurate operation.

Charge Pump: It converts a digital error signal generated by the phase detector into an analog control voltage that is used to adjust the frequency and phase of the PLL's output signal.

Loop Filter: It is used to filter and shape the analog control voltage generated by the charge pump. The loop filter helps to control the response of the PLL, shaping its frequency response and determining its stability, settling time, and overall performance.

Voltage-Controlled Oscillator: A voltage-controlled oscillator (VCO) will be included in the clock generator circuit to provide frequency tuning capability. The VCO will be controlled by a control voltage, which can be adjusted to vary the output frequency of the clock signal within the desired frequency range.

CML to CMOS Comparator : The output voltage swing at the output of VCO at such high frequencies might not be enough to drive digital logic in a programmable frequency divider. This block converts them into CMOS logic which can comfortably be driven by the digital blocks ahead of it.

Programmable Frequency Divider: A frequency divider circuit will be used to divide the output frequency of the VCO which will then be compared at PFD. The frequency divider will be designed to provide accurate frequency division with minimal jitter and frequency error.

Delta Sigma Modulator: This block is explicitly present in Fractional PLLs which are used to implement the fractional frequency division (FDF) or fractional frequency multiplication (FFM) which gives us a level more flexibility to give frequencies for Analog and Digital Blocks on chip whom to be tested

Digital Logic Circuitry: Wide range of clock frequencies desired by analog and digital blocks on the chip is generated from PLL with the help of digital logic circuitry.

## III. Summary of your Idea

The main specifications in the design of a PLL are phase noise and jitter In general, the phase noise and jitter performance of a PLL will depend on factors such as the loop bandwidth, phase detector type, VCO gain and tuning range, and the quality of the reference clock. Power consumption will also depend on the specific implementation here we plan to design an fractional PLL which will be a mixed signal block where it will have Digital PFD and programmable control logic for the selection of  suitable divider ratio  where a DCO or VCO architectures can be chosen based on other limitations on inductors and capacitors and area available and other factors .
The main factor responsible for phase noise filtering is loop bandwidth . A loop bandwidth of 2.5 - 5.5 MHz should be narrow enough to provide good phase noise across our desired offset frequencies and then slightly adjust it based on the settling time required to meet its specification .
The use of D-Flip Flop type may also impact the phase noise and jitter performance as this type of PFD can introduce dead zones and other non-linearities that may increase the overall jitter of the system.
VCO with gain (Kvco) of 20 MHz/v to 150 MHz/v is necessary to operate in GHz range so based on the desired frequency and other specifications which might be governed by other blocks we will need to choose a suitable VCO which is one of the critical block of a PLL .A Ring VCO has an advantage of good tuning flexibility with little area but poor phase noise performance whereas an LC based oscillator has good phase noise performance but poor tuning and large area .
The other small but important factors such as output frequency resolution , Lock time and Spurious suppression will be planned accordingly based on the blocks which will use this output from the PLL 
Charge pump with a programmable loop filter enables the loop bandwidth and phase noise to be optimized for the specific application.This is one of the control parameter where we choose different loop filters based on different parameters and frequency range and things like that so that more flexibility is given to user who tests an ASIC with this lab bench on a chip .
We are open to any architecture provided we take care of the specifications to be met and its ability to give flexibility for the user to use it and change it according to his needs in a lab bench on a chip , Having said that each Architecture come with its own limitations as well as advantages and implementation methods so we will choose suitable one so as to meet the desired specifications and we plan to have the below mentioned specifications .
So we would propose to design the Clock generator which is functional block in a Lab bench on chip our group has already taped out a PLL for sensor applications but now we would like to use those skills and enhance them further in  designing a compact and flexible Clock Generator block so as to be a  perfect fit to drive ADC’s , DAC’s and other Custom Analog or Digital Blocks with good phase noise and jitter performance .
