# **Introduction**

Programmable Gain Amplifier is used to deliver  to the ADC its required input level, Gain control can be achieved by either discrete steps (PGA) or by continuous gain variations (VGA).

A PGA is not used alone its a part of Automatic Gain Control loop, AGC could be classified into two categories, According to the path of the sensed signal (Feedback loop or Feedforward loop),Also according to the type of control (Digital or Analogue Control).

In our project a cell based technique will be used to design the PGA. The cell based technique achieves gain variation by cascading an optimized unit cell, these unit cells are controlled digitally by switches.

In this project two optimized unit cells will be designed to achieve the programmable gain specifications. Hence Feedforward AGC will be chosen, to avoid instability issues, As the PGA operates in an open loop configuration.

AGC loop consists of 3 main blocks, Variable Gain Amplifier, Peak Detector and Gain Control Voltage Generation Circuit.

# **PGA Architecture**

PGA is the core cell in the AGC system. PGA adjusts (amplifying or attenuating) the incoming signal to the desired optimal level of the ADC according to the control signal provided by the AGC loop.

A cell based technique is adopted into our project, which is achieved by cascading an pair of optimized unit cells (As shown in the following Figure), to get the required gain range.

Screenshot_1.png

The circuit consists of a basic differential amplifier with active loads, M1 and M2 are the input differential pair, M3 and M4 are a diode connected load, M5 and M6 are current sources, M7 is the tail current source. The gain of the topology mentioned above is the ratio between the transconductance of M1 to the transconductance of M3.

# **Peak Detector Architecture**

Peak Detectors are a key block in gain control, they detect the amplitude of the incoming signal and track this value throughout time, The selection of a feedforward control structure considerably increase the peak detector's performance requirements, since its input dynamic range and the linearity 
demand is enlarged.

The topology chosen is the OP-amp plus source follower based peak detector as shown in the following figure.


Peak Detector.png

The topology consists of an OP-amp followed by a source follower and a hold capacitor. OP-amp can be employed in feedback configuration so that the output of the source follower is connected to the op-amp negative input.

The circuit operates as follows, while Vin exceeds Vpeak M1 is on which charges the Capacitor C, While Vin goes below Vpeak M1 is off, and the capacitor holds the output peak voltage.

# **Gain Control Voltage Generation Circuit Architecture**

The control voltage generator takes the output signal from the peak detector and compares it with a reference signal then generate the control signal VC required to adjust the gain in the VGA.

There are two types of control digital and analogue, in this design digital control is chosen as it can provide acceptable gain steps transitions.

This control can be achieved by a comparator bank cell as shown in the following figure.
