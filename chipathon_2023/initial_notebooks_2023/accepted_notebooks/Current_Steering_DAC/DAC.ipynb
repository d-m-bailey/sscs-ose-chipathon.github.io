# **Introduction**
The DAC architectures are categorized into three main types according to the unit element that is used to construct the DAC circuit such as capacitive, resistive, and current DACs. The output impedance of resistive and capacitive DACs varies with the digital input code, so a buffer is needed to drive the next circuits, which limits the DACâ€™s maximum sampling rate. In addition, targeting high resolution using these types requires a large area to achieve good matching between the unit elements, which leads to large output capacitance and resistance. consequently, this degrades the settling time at the DAC output and limits the speed of the first two types.

On the other hand, current steering DAC can directly drive a small load termination without introducing an additional buffer, and also the steering behavior of the circuit is inherently fast. Accordingly, current steering DAC is considered an optimum choice to achieve high-speed and high-resolution design.

A General block diagram of our current steering DAC with a cascode unit current cell is shown in the figure. The number of uit current cells is dictated by the required number of bits. 

Segmented current steering DAC architecture is considered the best choice for designing a high resolution with high linearity DAC as it benefits from the advantage of combining both unary and binary-weighted unit cells. The unary unit cell has a smaller differential non-linearity (DNL), monotonicity, and mismatch requirement which reflects on a small area of the current source while the binary-weighted cell on the other hand has a small number of the unit cells and switches that reflect on design compactness and reduce routing complexity.

In our project, the target is to design a current steering DAC with 10-bit resolution, segmented into 6 thermometer encoded MSBs, and 4 binary weighted LSBs.


*   **Differential Pair:**
Steers the current to the right or left according to the digital input. 

*   **Cascode Current Source:**
Provides the constant DC Current that is to be steered between the two transistors of the differential pair. The cascode structure was chosen to provide high output impedance to reduce the current dependency on the DAC's output voltage.

*   **Resistors:**
Current to Voltage converters.

*   **Decoder/Dummy:**
The MSBs digital inputs are applied to a decoder that converts them into thermometer signals that are compatible with the input of unary DAC. On the other hand, LSBs digital inputs are passed through a dummy decoder to equalize the delay resulting from the thermometer decoder in the unary chain.
